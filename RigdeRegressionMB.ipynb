```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = make_regression(n_samples=100, n_features=1,n_informative=1, n_targets=1, noise=20, random_state=13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha = 10)
rr.fit(x,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1 = Ridge(alpha = 200)
rr1.fit(x,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
plt.plot(x,y,'b.')
plt.plot(x,lr.predict(x),color = 'red',label = 'alpha=0')
plt.plot(x,rr.predict(x),color = 'green',label = 'alpha=10')
plt.plot(x,rr1.predict(x),color = 'blue',label = 'alpha=200')
plt.legend()
```

```python
class OneDRidgeRegression:
    def __init__(self,alpha =0.1):
        self.alpha = alpha
        self.m = None
        self.b = None
    
    def fit(self,x_train,y_train):
        num = 0
        den = 0

        for i in range(x_train.shape[0]):
            num = num + (y_train[i] - y_train.mean())*(x_train[i] - x_train.mean())
            den = den + (x_train[i] - x_train.mean())*(x_train[i] - x_train.mean())
        self.m = num/(den + self.alpha)
        self.b = y_train.mean() - (self.m * x_train.mean())
        print(self.m,self.b)

    def predict(self,x_test):
        return x_test.self.m + self.b
```

```python
ridge = OneDRidgeRegression(alpha = 200)
```

```python
print(ridge.fit(x,y)) # our algorithm
print(rr1.coef_,rr1.intercept_) # sklearn algorithm
```
